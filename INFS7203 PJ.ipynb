```python
import pandas as pd
```

```python
df = pd.read_csv('Ecoli.csv')
```

```python
class_0 = df.loc[df['Target (Col 107)']==0]
```

```python
class_1 = df.loc[df['Target (Col 107)']==1]
```

```python
df.columns
```

```python
for i in df.columns:
    if 'Nom' in i:
        class_0[i].fillna((class_0[i].mode()), inplace=True)
        class_1[i].fillna((class_1[i].mode()), inplace=True)
    if 'Target' in i:
        continue
    else:
        class_0[i].fillna((class_0[i].mean()), inplace=True)
        class_1[i].fillna((class_1[i].mean()), inplace=True)
```

```python
process_df = pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
process_df
```

```python
X = process_df.drop('Target (Col 107)',axis=1)
y = process_df['Target (Col 107)']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(X,y,random_state=13, test_size=0.20,stratify=y,shuffle=True)
```

# Density Based Outlier Detection

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor(n_neighbors=10)
pred_outlier = lof.fit_predict(X)
```

```python
density_based_process_df = process_df.copy()
density_based_process_df['outlier_flag'] = pred_outlier 
```

```python
density_based_process_df = density_based_process_df.loc[density_based_process_df['outlier_flag'] == 1]
```

```python
density_X = density_based_process_df.drop(['Target (Col 107)','outlier_flag'],axis=1)
density_y = density_based_process_df['Target (Col 107)']
```

```python
X_train_density, X_test_density,y_train_density, y_test_density = train_test_split(density_X,density_y,random_state=13,
                                                                                   test_size=0.20,stratify=density_y,shuffle=True)
```

## Min-Max Norm

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
density_minmax_X = min_max_scaler.fit_transform(X_train_density)
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler = StandardScaler()
density_std_X = std_scaler.fit_transform(X_train_density)
```

# 10-Fold cross validation for pre-processing

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

d_tree = DecisionTreeClassifier(random_state=13)
gnb = GaussianNB()
```

```python
def get_10fcv_score(model,X,y,word):
    scores = cross_val_score(model,X, y, cv=10,scoring='f1_macro')
    return print('Average F1 score of ', word, ':',round(scores.mean(),2))
```

## Decision Tree

```python
# Contain Outliers
get_10fcv_score(d_tree,X_train,y_train,'No Processing')
get_10fcv_score(d_tree,base_min_max_X,y_train,'Min-Max Only')
get_10fcv_score(d_tree,base_std_X,y_train,'Standardize Only')

# Density-based Detection
get_10fcv_score(d_tree,X_train_density,y_train_density,'Density-based')
get_10fcv_score(d_tree,density_minmax_X,y_train_density,'Density-based - Min-Max')
get_10fcv_score(d_tree,density_std_X,y_train_density,'Density-based - Standardize')

# Isolation-based Detection
get_10fcv_score(d_tree,X_train_iso,y_train_iso,'Isolation-based')
get_10fcv_score(d_tree,iso_minmax_X,y_train_iso,'Isolation-based Min-Max')
get_10fcv_score(d_tree,iso_std_X,y_train_iso,'Isolation-based Standardize')
```

From the 10-fold cross validation with decision tree, we have found that the density-based outlier detection perform better in average F1 scores than not removing the outliers and remove the outliers with isolated-based method.

Next Step, I am going to find which normalization method with density-based detection by using Naive Bayes as learner.

# 10-fold Cross Validation for learner selection

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, VotingClassifier

d_tree = DecisionTreeClassifier(random_state=13)
gnb = GaussianNB()
knn = KNeighborsClassifier(
) # default at 5
rf = RandomForestClassifier(random_state=13)
emb_vote = VotingClassifier(estimators=[('gnb', gnb), ('knn', knn), ('d-tree', d_tree)])
```

```python
get_10fcv_score(d_tree,X_train_density,y_train_density,'Decision Tree')
get_10fcv_score(gnb,X_train_density,y_train_density,'Navie Bayes')
get_10fcv_score(knn,X_train_density,y_train_density,'K-NN')
get_10fcv_score(rf,X_train_density,y_train_density,'Random Forest')
get_10fcv_score(emb_vote,X_train_density,y_train_density,'Emsemble')
```

From the 10-fold cross validation for the learners with sklearn default setting, decision tree and the ensemble learner with Naive Bayes, K Nearest Neighbors and decision tree have highest average score on F1. Thus, we are going to evaluate this two learners with balanced accuracy metrics. 

```python
scores = cross_val_score(d_tree,X_train_density, y_train_density, cv=10,scoring='balanced_accuracy')
print('Average accuracy score of ', 'Dtree', ':',round(scores.mean(),2))
```

```python
scores = cross_val_score(emb_vote,X_train_density, y_train_density, cv=10,scoring='balanced_accuracy')
print('Average accuracy score of ', 'Emsemble', ':',round(scores.mean(),2))
```

From the average balanced accuracy, we can find that the emsemble learner has a slightly higher accuracy but not much differrence. The emsemble model has increase the complexity while not improving much in the prediction power. Thus, I will choose decision tree as the learner to be used and optimized.

# Model Fine Tuning

In this section, I will fine tune the hyperparameter of decision tree and validate with the test set

```python
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
```

```python
def d_tree_cv_tune(measure,split,depth,node):
    d_tree = DecisionTreeClassifier(random_state=13, 
                                    criterion=measure,splitter = split, max_depth = depth,max_leaf_nodes=node)
    score = cross_val_score(d_tree,X_train_density, y_train_density, cv=10,scoring='f1_macro')
    return score.mean()
```

```python
d_tree_cv_tune('entropy','random',4,None)
```

```python
d_tree_cv_tune('gini','random',3,None)
```

```python
d_tree_cv_tune('gini','best',1,None)
```

```python
d_tree_cv_tune('entropy','best',2,None)
```

```python
def d_tree_tuning(X_train_density,y_train_density,X_test_density,y_test_density,measure,split,depth):
    d_tree = DecisionTreeClassifier(random_state=13, 
                                    criterion=measure,splitter = split, max_depth = depth)
    d_tree.fit(X_train_density,y_train_density)
    pred_y = d_tree.predict(X_test_density)
    
    print('Accuracy:',balanced_accuracy_score(y_test_density, pred_y))
    
    conf_mat = pd.DataFrame(confusion_matrix(y_test_density, pred_y))
    conf_mat.columns =['Predicted Positive', 'Predicted Negative']
    conf_mat.index = ['True Positive', 'True Negative']
    
    print('Precision:',precision_score(y_test_density, pred_y,average='macro'))
    print('Recall:',recall_score(y_test_density, pred_y,average='macro'))
    print('F1 Score:',f1_score(y_test_density, pred_y,average='macro'))
    print('------------------------------------------------------------')
    
    print(conf_mat)
    
    return d_tree
```

```python
d_tree_tuning(X_train_density,y_train_density,X_test_density,y_test_density,'gini','best',None,None)
```

```python
d_tree_tuning(X_train_density,y_train_density,X_test_density,y_test_density,'entropy','best',None,None)
```

```python
# Best condition for each case
d_tree_tuning(X_train_density,y_train_density,X_test_density,y_test_density,'gini','best',1,None)
d_tree_tuning(X_train_density,y_train_density,X_test_density,y_test_density,'entropy','best',2,None)
d_tree_tuning(X_train_density,y_train_density,X_test_density,y_test_density,'entropy','random',4,None)
d_tree_tuning(X_train_density,y_train_density,X_test_density,y_test_density,'gini','random',3,None)
```

```python
model = d_tree_tuning(X_train_density,y_train_density,X_test_density,y_test_density,'entropy','random',4)
```

Best case is entropy with random split and max_depth is 4
